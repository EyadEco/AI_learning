```python
import numpy as np
import pandas as pd
import re
from vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer
import transformers 
from transformers import pipeline
```

```python
data = pd.read_csv('book_reviews_sample.csv')
data.head()
```

```python
data["reviewText_clean"] = data["reviewText"].apply(lambda x : re.sub(r"([^\w\s])" , "" , x)).str.lower()
```

```python
data.head()
```

```python
vaderSentiment = SentimentIntensityAnalyzer()
```

```python
data["vader_sentiment_score"] = data['reviewText_clean'].apply(lambda x : vaderSentiment.polarity_scores(x)['compound'])
```

```python
data.head()
```

```python
bin = [-1 , -0.1 , 0.1 , 1]
names = ['negative' , 'neutral' , 'positive']
data['vander_statment_label'] = pd.cut(data["vader_sentiment_score"] , bin , labels=names)
```

```python
data['vander_statment_label'].value_counts().plot.bar()
```

```python
transformer_pipeline = pipeline("sentiment-analysis")
```

```python
data["transform_sentiment_label"]= data['reviewText_clean'].apply(lambda x : transformer_pipeline(x)[0]["label"])
```

```python
data['transform_sentiment_label'].value_counts().plot.bar()
```

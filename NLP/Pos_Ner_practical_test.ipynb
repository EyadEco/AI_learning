```python
import nltk
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
from nltk.stem import PorterStemmer , WordNetLemmatizer
import pandas as pd
import re
import matplotlib.pyplot as plt
import spacy
```

```python
bbc_data = pd.read_csv("bbc_news.csv")
```

```python
bbc_data.head()
```

```python
bbc_data_titles = pd.DataFrame(bbc_data['title'])
```

```python
print(type(bbc_data_titles))
```

```python
bbc_data_titles.head()
```

```python
bbc_data_titles["lower_case_title"] = bbc_data_titles['title'].str.lower()
```

```python
bbc_data_titles.head()
```

```python
en_stopwords = stopwords.words('english')
```

```python
bbc_data_titles['titles_no_stopwords'] = bbc_data_titles['title'].apply(lambda x: ' '.join([word for word in x.split() if word not in en_stopwords]))
```

```python
bbc_data_titles["title_no_stopwords_no_punct"] = bbc_data_titles['titles_no_stopwords'].apply(lambda x : re.sub(r"[^\w\s]" , "" , x))
```

```python
bbc_data_titles["token_raw"] = bbc_data_titles['title'].apply(lambda x: word_tokenize(x))
```

```python
bbc_data_titles["token_clean"] = bbc_data_titles['title_no_stopwords_no_punct'].apply(lambda x: word_tokenize(x))
```

```python
bbc_data_titles.head()
```

```python
lemmatizer = WordNetLemmatizer()
```

```python
bbc_data_titles["token_clean_lemmatized"] = bbc_data_titles['token_clean'].apply(lambda x : [lemmatizer.lemmatize(word)for word in x])
```

```python
token_raw_list = sum(bbc_data_titles['token_raw'] , [])
token_clean_list = sum(bbc_data_titles['token_clean_lemmatized'] , [])
```

pos

```python
nlp = spacy.load('en_core_web_sm')
```

```python
doc = nlp(' '.join(token_raw_list))
```

```python
pos_df = pd.DataFrame(columns=['token' , 'pos_tag'])
```

```python
for token in doc:
    pos_df = pd.concat([pos_df , pd.DataFrame.from_records([{'token' : token.text , 'pos_tag' : token.pos_}])] ,ignore_index=True)
```

```python
pos_df_counts = pos_df.groupby(['token' , 'pos_tag']).size().reset_index(name='counts').sort_values(by='counts',ascending=False)
pos_df_counts.head(10)
```

```python
noun = pos_df_counts[pos_df_counts.pos_tag == 'NOUN'][:10]
noun
```

```python
verb = pos_df_counts[pos_df_counts.pos_tag == 'VERB'][:10]
verb
```

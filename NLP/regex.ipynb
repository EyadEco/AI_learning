```python
customer_reviews=["sam wad a great help to me in the store",
                  "the cashier wad very rude to me, I think her name was eleanor",
                  "sarh was able to help me find the items I needed quickly",
                  "lucy is such a great addition to the team",
                  "a great service from sarah she found me what i wanted"]
```

```python
sarahs_reviews=[]
```

```python
pattern=r"sara?h"
```

```python
for review in customer_reviews:
    if re.search(pattern , review):
        sarahs_reviews.append(review)

print(sarahs_reviews)
```

```python
a_pattern = r"^a"
a_reviews = []
for review in customer_reviews:
    if re.search(a_pattern , review):
        a_reviews.append(review)

print(a_reviews)
```

```python
y_reviews = []
y_pattern=r"y$"
for review in customer_reviews:
    if re.search(y_pattern , review):
        y_reviews.append(review)

print(y_reviews)
```

```python
want_need = []
wantneed_pattern=r"(need|want)ed"
for review in customer_reviews:
    if re.search(wantneed_pattern , review):
        want_need.append(review)

print(want_need)
```

```python
no_punct_reviews = []
no_punct_pattern=r"[^\w\s]"
for review in customer_reviews:
    new = re.sub(no_punct_pattern , "" , review)
    no_punct_reviews.append(new)

print(no_punct_reviews)
```

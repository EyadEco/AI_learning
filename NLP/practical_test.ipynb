```python
import nltk
from nltk.tokenize import word_tokenize
from nltk.stem import PorterStemmer , WordNetLemmatizer
from nltk.corpus import stopwords
import re
import pandas as pd
```

```python
data = pd.read_csv("./tripadvisor_hotel_reviews.csv")
```

```python
data.info()
```

```python
data.head()
```

```python
data["review_lower"] = data["Review"].str.lower()
```

```python
data.head()
```

```python
en_stopwords = stopwords.words("english")
```

```python
en_stopwords.remove("not")
```

```python
data["review_no_stopwords"] = data["review_lower"].apply(lambda x : ' '.join([word for word in x.split() if word not in en_stopwords]))
```

```python
data.head()
```

```python
data["review_no_punct_no_stopword"] = data.apply(lambda x : re.sub(r"[*]" , "star" , x["review_no_stopwords"]), axis=1)
```

```python
data.head()
```

```python
data["review_no_punct_no_stopword"] = data.apply(lambda x : re.sub(r"[^\w\s]" , "" , x["review_no_punct_no_stopword"]), axis=1)
```

```python
data["tokenization"] = data.apply(lambda x : word_tokenize(x["review_no_punct_no_stopword"]) , axis=1)
```

```python
data["tokenization"][0]
```

```python
ps = PorterStemmer()
```

```python
data["stemmed"] = data.apply(lambda x : [ps.stem(word) for word in x["tokenization"]] , axis=1)
```

```python
data["stemmed"]
```

```python
lemmatizer = WordNetLemmatizer()
```

```python
data["lemmatized"] = data.apply(lambda x : [lemmatizer.lemmatize(word) for word in x["tokenization"]] , axis=1)
```

```python
token_clean = sum(data["lemmatized"] , [])
```

```python
unigram = (pd.Series(nltk.ngrams(token_clean, 1)).value_counts())
print(unigram)
```

```python
bigram = (pd.Series(nltk.ngrams(token_clean, 2)).value_counts())
print(bigram)
```
